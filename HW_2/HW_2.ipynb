## Problem 1

```
from mpl_toolkits import mplot3d

x = np.linspace(-1, 1, 100)
y = np.linspace(-1, 1, 100)

X, Y = np.meshgrid(x, y)
```

$f_1(x, y) = x^2 + y^2 \,\,\,\,\,\,\,\,\,\,\,\,\,\, p_1 = [0, 0, 0]$    


```
def f1(x, y):
    return np.power(x ** 2 + y ** 2, 1)

def p1():
    return [0, 0]
```

$f_2(x, y) = -x^2 - y^2 \,\,\,\,\,\,\,\,\, p_2 = [0, 0, 0]$    

```
def f2(x, y):
    return np.power(- x ** 2 - y ** 2, 1)

def p2():
    return [0, 0]
```

$f_3(x, y) = x^2 - y^2 \,\,\,\,\,\,\,\,\,\,\,\,\,\, p_3 = [0, 0, 0]$    

```
def f3(x, y):
    return np.power(x ** 2 - y ** 2, 1)

def p3():
    return [0, 0]
```

## Problem 2

```
import numpy as np
%matplotlib inline
import matplotlib as mpl
import matplotlib.pyplot as plt
mpl.rc('axes', labelsize=14)
mpl.rc('xtick', labelsize=12)
mpl.rc('ytick', labelsize=12)
```

```
# generate m random pairs (x, y)
m = 100
X = 2 * np.random.rand(m, 1)
X1 = 2 * np.random.rand(m, 1)
X2 = 2 * np.random.rand(m, 1)
X3 = 2 * np.random.rand(m, 1)
y = 4 + 3 * X + 5*X1 + np.random.rand(m, 1)
y1 = 4 + 3 * X2 + 5*X3 + np.random.rand(m, 1)
```

```
plt.plot(X, y, "b.")
plt.xlabel("$x_1$", fontsize=18)
plt.ylabel("$y$", rotation=0, fontsize=18)
plt.axis([0, 2, 0, 15])
plt.show()
```

```
# number of epochs
epochs = 20
# learning rate
lr = 0.01
# fix initial random weight for gradient descent (batch gradient descent)
np.random.seed(42)
initial_weight = np.random.randn(3, 1)

# add x0 = 1 
X_b = np.column_stack([np.ones((m, 1)), X, X1])
```

```
weight = initial_weight
weight_path_mgd = []
batch_size = 4

weight_path_mgd.append(weight)
for epoch in range(epochs):
    shuffled_indices = np.random.permutation(m)
    X_b_shuffled = X_b[shuffled_indices]
    y_shuffled = y[shuffled_indices]
    for i in range(0, m, batch_size):
        xi = X_b_shuffled[i:i+batch_size]
        yi = y_shuffled[i:i+batch_size]
        gradient = 1 / batch_size * xi.T.dot(xi.dot(weight) - yi)
        weight = weight - lr * gradient
        weight_path_mgd.append(weight)

weight_path_mgd = np.array(weight_path_mgd)
y_hat = weight_path_mgd[-1][2]*X3 + weight_path_mgd[-1][1]*X2 + weight_path_mgd[-1][0]
```

```
ax = plt.axes(projection='3d')

ax.scatter3D(X2, X3, y_hat)
ax.scatter3D(X2, X3, y1)
```

## Problem 3

```
%tensorflow_version 2.x
import tensorflow as tf

(train_images, train_labels), (test_images, test_labels) = tf.keras.datasets.mnist.load_data()
```

```
digit_template = np.zeros((28, 28))
digit_array = np.array([np.copy(digit_template), np.copy(digit_template), np.copy(digit_template), np.copy(digit_template), np.copy(digit_template), np.copy(digit_template), np.copy(digit_template), np.copy(digit_template), np.copy(digit_template), np.copy(digit_template)])

for n in range(10):
    digit_count = 0
    for i in range(len(train_labels)):
        if (train_labels[i] == n):
            digit_array[n] = digit_array[n]+train_images[i]
            digit_count = digit_count+1
    
    digit_array[n] = digit_array[n]/digit_count

plt.figure(figsize=(16, 16))
for n in range(10):
    ax = plt.subplot(4, 4, n + 1)
    plt.imshow(digit_array[n])
    plt.title(n)
    plt.axis('off')
```
